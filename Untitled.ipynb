```python
from moviepy.editor import *
import imageio
from matplotlib.pyplot import imshow
import numpy as np
from skimage import color
from skimage import io
import pytesseract
#tessdata_dir_config= '--tessdata-dir "C:\Program Files (x86)\Tesseract-OCR\tesseract"'
from PIL import Image
import itertools
```

```python
TRESHOLD=220
FREQUENCY=2
```

```python
hqclip =  VideoFileClip("hqclip.mp4")
```

```python
mask = color.rgb2gray(imageio.imread('mask.png'))
mask_ocr = color.rgb2gray(imageio.imread('mask_ocr.png'))
```

```python
%matplotlib inline
imshow(np.asarray(mask), cmap='prism')
mask=mask//240
```

```python
imshow(np.asarray(mask_ocr), cmap='Greys')
mask_ocr=mask_ocr//240
```

```python
frames = (int)(hqclip.duration*hqclip.fps)
frame= (color.rgb2gray(hqclip.get_frame(t=220))*256).astype(int)
```

```python
def parseFrame(time):
    frame= (color.rgb2gray(hqclip.get_frame(t=time))*256).astype(int)
    frame_detect=frame*mask
    frame_masked=np.ma.masked_equal(frame_detect, 0)
    if(np.ma.mean(frame_masked)<TRESHOLD): return False
    return frame*mask_ocr
```

```python
def trimFrameIntoOCRArray(frame):
    arr=np.asarray(frame)
    arr= arr[~np.all(arr == 0, axis=1)]
    arr= arr[:, ~np.all(arr == 0, axis=0)]
    #imshow(arr, cmap='Greys')
    return arr
```

```python
frame=parseFrame(220)
```

```python
def ocr(frame):
    return pytesseract.image_to_string(Image.fromarray(frame))
```

```python
hqclip.ipython_display(audio=False,width=200,t=3315)
#imshow(trimFrameIntoOCRArray(parseFrame(3315)))
```

```python
def LookForText(i): 
    frame_maybe = parseFrame(i/hqclip.fps)
    if(frame_maybe is not False):
        frame=trimFrameIntoOCRArray(frame_maybe)
        result=ocr(frame)
        return result
    return False
```

```python
def parseQuestion(text):
    if(not text): return False
    lines=text.splitlines()
    text=list(filter(lambda x:x.strip() is not "" ,[' '.join(list(y)) for x, y in itertools.groupby(lines, lambda z: z == "") if not x]))
    if(len(text)<4): return []
    return text
```

```python
parseQuestion(LookForText(3165))
```

```python
for i in range(3000,frames,(int)(hqclip.fps/FREQUENCY)):
    print(i,parseQuestion(LookForText(i)))
```
